```python
import deepdoctection as dd
import pandas as pd
import ast
from copy import deepcopy
```

```python
csv_path = "/home/janis/Tests/scrapy/data.csv"
```

```python
df = pd.read_csv(csv_path)
df_list_dict = df.to_dict("records")
len(df_list_dict)
```

```python
def tear_down_files(cell):
    cell_dict = ast.literal_eval(cell)[0]
    return cell_dict
```

```python
all_items=[]
for row, row_ele in zip(df[["files"]].iterrows(),df_list_dict):
    row_ele.pop("files")
    files_dict = tear_down_files(row[1]["files"])
    row_ele["url"]= files_dict["url"]
    row_ele["path"]= files_dict["path"]
    all_items.append(row_ele)
```

```python
for idx, row in enumerate(all_items):
    if isinstance(row["author"], float):
        row["author"] = [[]]
    else:
        authors = row["author"].split(",")
        all_authors = []
        for idx, name in enumerate(authors):
            if idx % 2 == 0:
                author = []
                author.append(name)
            else:
                author.append(name)
                all_authors.append(author)
            row["author"] = all_authors            
```

```python
all_items_per_author = []

for row in all_items:
    authors = row.pop("author")
    for author in authors:
        cp_row = deepcopy(row)
        cp_row["author"] = author
        all_items_per_author.append(cp_row)
```

```python
df = pd.DataFrame(all_items_per_author)
```
