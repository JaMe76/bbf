```python
import deepdoctection as dd
from PyPDF2 import PdfFileWriter, PdfFileReader
from pathlib import Path
import os
import jsonlines
```

```python
def reverse_pdf(path):
    output_pdf = PdfFileWriter()

    with open(path, 'rb') as readfile:
        path = Path(path)
        tmp_file_name = path.name.replace(".pdf","_tmp.pdf")
        input_pdf = PdfFileReader(readfile)
        total_pages = input_pdf.getNumPages()
        for page in range(total_pages - 1, -1, -1):
            output_pdf.addPage(input_pdf.getPage(page))
        tmp_path = (path.parent / tmp_file_name).as_posix()
        with open(tmp_path, "wb") as writefile:
            output_pdf.write(writefile)
    return tmp_path, total_pages
```

```python
base_path = Path("/home/janis/Data/full_cut_files")
df = dd.SerializerFiles.load(path=base_path, file_type=".pdf")
df.reset_state()

for idx, dp in enumerate(df):
    file_name = Path(dp).name
    print(f"{idx}: processing {file_name}")
    path, _ = reverse_pdf(dp)
    os.rename(path, base_path / file_name)
```

```python
def get_layout_analyzer():

    layout_config_path = (dd.get_configs_dir_path() / "dd/d2/layout/CASCADE_RCNN_R_50_FPN_GN_nms_higher.yaml").as_posix()
    layout_weights_path = "/home/janis/Experiments/fine_tune_xrf_d2/pdocs_170223_2/model_0000199.pth"
    categories_layout={
                "1": dd.LayoutType.text,
                "2": dd.LayoutType.title,
                "3": dd.LayoutType.list,
                "4": dd.LayoutType.table,
                "5": dd.LayoutType.figure,
            }
    config_overwrite = [f"NMS_THRESH_CLASS_AGNOSTIC=0.2",
                        f"MODEL.ROI_HEADS.SCORE_THRESH_TEST=0.068",
                        f"MODEL.ROI_HEADS.NMS_THRESH_TEST=0.2"]
    d_layout = dd.D2FrcnnDetector(layout_config_path, layout_weights_path, categories_layout, config_overwrite, device="cuda")
    layout = dd.ImageLayoutService(d_layout, to_image=True,)
    tess_ocr_config_path = "/home/janis/Repos/bbf/pedocs/conf_tesseract.yaml"
    d_tess_ocr = dd.TesseractOcrDetector(tess_ocr_config_path)
    text = dd.TextExtractionService(d_tess_ocr)

    match = dd.MatchingService(
        parent_categories=[dd.LayoutType.text,
                           dd.LayoutType.title,
                           dd.LayoutType.list,
                           dd.LayoutType.table,
                           dd.LayoutType.figure],
        child_categories=dd.LayoutType.word,
        matching_rule="ioa",
        threshold=0.4,
    )

    order = dd.TextOrderService(
        text_container=dd.LayoutType.word,
        floating_text_block_names=[dd.LayoutType.title, dd.LayoutType.text, dd.LayoutType.list],
        text_block_names=[
            dd.LayoutType.title,
            dd.LayoutType.text,
            dd.LayoutType.list,
            dd.LayoutType.cell,
            dd.CellType.header,
            dd.CellType.body,
        ],
    )


    pipe = dd.DoctectionPipe(pipeline_component_list=[layout,text,match,order])

    return pipe
```

```python
def parse_document(path, analyzer):
    meta_data_dir = Path("/home/janis/Data/pedocs_meta_data")
    document_id = Path(path).stem
    
    df = analyzer.analyze(path=path)
    df.reset_state()
    all_text_items = []
    all_images = []
    
    extract_from_block = 0
    start_from_page = 0
    trigger_found = False
    
    for page_num, dp in enumerate(df):
        for layout in dp.layouts:
            if layout.category_name == "title":
                if layout.text in ("Literatur",
                                   "Literatur (seit 1945)",
                                   "Sekundärliteratur",
                                   "References",
                                   "Literatur:",
                                   "Bibliography",
                                   "Literatur und Quellen",
                                   "Gedruckte Quellen und Literatur",
                                   "Weiterführende Literatur",
                                   "Literaturverzeichnis",
                                   "Literature",
                                   "Formatierung",
                                   "5. Literatur",
                                   "b) Literatur"):
                    start_from_page = page_num
                    extract_from_block = layout.reading_order
                    trigger_found= True
        if trigger_found:
            all_images.append(dp)
            dp.save(path=meta_data_dir / dp.file_name)
            
    print(f"Extracting from {document_id}, starting at page {start_from_page} \
    and extracting from block {extract_from_block}")
    
    if not trigger_found:
        all_text_items.append({"doc_id": document_id, "file_name": None, "annotation_id": None, "text": None, "page_num": None})
        return all_text_items
    
    for idx, dp in enumerate(all_images):
        if idx == 0:
            for layout in dp.layouts:
                if layout.category_name =="text" and layout.reading_order >=extract_from_block: #) or 
                    text_item = {"doc_id": document_id, "file_name": dp.file_name, "annotation_id":layout.annotation_id, "text": layout.text, "page_num": idx}
                    all_text_items.append(text_item)
        else:
            for layout in dp.layouts:
                if layout.category_name == "text":
                    text_item = {"doc_id": document_id, "file_name": dp.file_name, "annotation_id":layout.annotation_id, "text": layout.text, "page_num": idx}
                    all_text_items.append(text_item)
                                          
    return all_text_items

def launch(num_docs):
    dd.set_logger_dir("/home/janis/Data/pedocs_meta_data/logger")
    path = "/home/janis/Data/full_cut_files"
    df = dd.SerializerFiles.load(path=path, file_type=".pdf", max_datapoints=num_docs)
    analyzer = get_layout_analyzer()
    df.reset_state()
    
    with jsonlines.open('/home/janis/Data/pedocs_meta_data/raw_items/pedocs.jsonl', mode='a') as writer:
        for dp in df:
            parsed_text = parse_document(dp, analyzer)
            for element in parsed_text:
                writer.write(element)
```
